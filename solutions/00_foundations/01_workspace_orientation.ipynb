## Step 1: Create a Cluster

### What is a cluster?

A **cluster** is a set of cloud virtual machines (VMs) that work together to run your code. When you execute a cell in a Databricks notebook, the code is sent to the cluster, which distributes the work across its machines and returns the results to your screen.

Without a cluster, your notebook is just a text editor -- it cannot execute anything.

### Analogy

> A cluster is like **renting a kitchen**.
>
> - **Classic cluster** = You choose the stove size, number of burners, and gas type. It takes a few minutes to set up, but you get exactly what you need.
> - **Serverless** = Databricks gives you a fully equipped kitchen that appears in seconds. You do not choose the appliances -- they just work.

### How to Create a Classic Cluster (UI Walkthrough)

1. **UI -->** Left nav --> Click **Compute** (the lightning bolt icon)
2. Click the **Create compute** button (top right)
3. Configure the following:
   - **Cluster name:** `genai-labs` (or any name you prefer)
   - **Cluster mode:** Single Node (sufficient for these labs; saves cost)
   - **Databricks Runtime:** Choose the latest **LTS** (Long Term Support) version
   - **Node type:** Smallest available (e.g., `Standard_DS3_v2` on Azure, `i3.xlarge` on AWS)
   - **Auto termination:** Set to **60 minutes** (the cluster shuts down after 60 min of inactivity to save money)
4. Click **Create Compute** and wait 3--7 minutes for it to start

### Serverless Alternative

If your workspace supports serverless compute:
- Simply open a notebook and select **Serverless** from the compute dropdown at the top
- No configuration needed -- it starts in seconds
- Ideal for quick exploration; classic clusters are better when you need custom libraries

> **Community Edition note:** Community Edition provides a free single-node cluster. Serverless is not available on Community Edition.

````python
# --------------------------------------------------------------------------
# Step 1 -- Check compute info programmatically
# --------------------------------------------------------------------------
# Works on both classic clusters AND serverless compute.

spark_version = spark.version
print(f"Spark version: {spark_version}")

try:
    cluster_name = spark.conf.get("spark.databricks.clusterUsageTags.clusterName")
    node_type = spark.conf.get("spark.databricks.clusterUsageTags.clusterNodeType", "unknown")
    print(f"Compute type : Classic cluster")
    print(f"Cluster name : {cluster_name}")
    print(f"Node type    : {node_type}")
except Exception:
    print(f"Compute type : Serverless")
    print(f"Cluster name : (managed by Databricks)")
    print(f"Node type    : (managed by Databricks)")

print(f"\nYou're connected and ready to go!")
````

## Step 3: Create a DataFrame and Write a Delta Table

### What is Delta Lake?

**Delta Lake** is an enhanced storage format built on top of **Parquet**. If you have worked with Parquet files before, you know they are fast and columnar -- great for analytics. Delta Lake adds several critical features on top:

- **ACID transactions** -- reads and writes are reliable, even when multiple jobs run at the same time
- **Versioning (time travel)** -- every change is recorded; you can query the table as it looked yesterday
- **Change Data Feed** -- you can see exactly which rows were inserted, updated, or deleted
- **Schema enforcement** -- the table rejects writes that do not match its schema

Delta Lake is the **default table format** on Databricks. Whenever you create a table, it is stored as Delta unless you specify otherwise.

### Analogy

> If Parquet is a **filing cabinet**, Delta Lake is a **filing cabinet with a change log** -- you can always see what was in there yesterday, who changed what, and roll back if something went wrong.

### What we will do

We will create a small in-memory DataFrame, then write it to the Unity Catalog as a Delta table at `workspace.genai_labs.sample_table`.

> **Note:** The three-part name `workspace.genai_labs.sample_table` follows the Unity Catalog naming convention: `catalog.schema.table`. We will cover this in depth in Lab 00-02.

````python
# --------------------------------------------------------------------------
# Step 3 -- Create a DataFrame and write it as a Delta table
# --------------------------------------------------------------------------

# First, make sure the schema exists (it may not on a fresh workspace)
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.genai_labs")

# Create a small sample DataFrame
data = [
    ("Alice", "Prompt Engineering", 92),
    ("Bob", "RAG Pipelines", 88),
    ("Carol", "Model Fine-Tuning", 95),
    ("David", "LLM Evaluation", 78),
    ("Eve", "Prompt Engineering", 85),
]

columns = ["student", "topic", "score"]

df = spark.createDataFrame(data, columns)

# Write to Delta format in Unity Catalog
df.write.format("delta").mode("overwrite").saveAsTable("workspace.genai_labs.sample_table")

print("Table created successfully: workspace.genai_labs.sample_table")
````

### Expected Output

```
Table created successfully: workspace.genai_labs.sample_table
```

If you see an error about the catalog or schema not existing, make sure Unity Catalog is enabled in your workspace and that you have permissions to write to `workspace.genai_labs`.

> **Community Edition note:** Community Edition does not support Unity Catalog. Replace the table path with a Hive metastore path like `default.sample_table` and remove the `CREATE SCHEMA` line.

## Step 4: Read the Delta Table Back

Now let's confirm the data was written correctly by reading it back from the catalog. We will show two ways to display the data:

1. **`df.show()`** -- prints a plain-text table to the cell output (works everywhere)
2. **`display(df)`** -- renders a rich, interactive table in Databricks notebooks (with sorting, filtering, and visualization options)

````python
# --------------------------------------------------------------------------
# Step 4 -- Read the Delta table back from the catalog
# --------------------------------------------------------------------------

# Method 1: .show() -- plain text output
print("=== Using .show() ===")
spark.table("workspace.genai_labs.sample_table").show()

# Method 2: display() -- rich Databricks rendering
print("=== Using display() ===")
display(spark.table("workspace.genai_labs.sample_table"))
````

## Step 5: Run a SQL Query

### Python vs. SQL in Databricks

Databricks notebooks support **multiple languages** in the same notebook. The default language is set at the top, but you can switch any individual cell:

- **`%sql`** -- Run SQL in a Python notebook (called a "magic command")
- **`%python`** -- Run Python in a SQL notebook
- **`%scala`**, **`%r`** -- Also available

Alternatively, you can click the **language dropdown** at the top-left corner of any cell and select SQL.

### When to use SQL vs. Python?

- **SQL** is excellent for quick data exploration, filtering, and aggregation. Most data analysts prefer it.
- **Python** (PySpark) is better for complex transformations, ML pipelines, and calling APIs.
- In practice, you will mix both freely. The exam expects you to be comfortable with either.

### Running SQL from Python

Below we use `spark.sql()` to run a SQL query from a Python cell. This is useful when you want to keep everything in Python but still leverage SQL syntax.

````python
# --------------------------------------------------------------------------
# Step 5 -- Run a SQL query from Python
# --------------------------------------------------------------------------

# Filter for students who scored above 85
result = spark.sql("""
    SELECT student, topic, score
    FROM workspace.genai_labs.sample_table
    WHERE score > 85
    ORDER BY score DESC
""")

result.show()
````

### Expected Output

```
+-------+------------------+-----+
|student|             topic|score|
+-------+------------------+-----+
|  Carol|  Model Fine-Tuning|   95|
|  Alice|Prompt Engineering|   92|
|    Bob|     RAG Pipelines|   88|
+-------+------------------+-----+
```

Only the three students with scores above 85 are returned, sorted from highest to lowest.

> **Equivalent `%sql` magic:** In a Databricks notebook, you could also create a new cell, type `%sql` at the top, and write the SQL directly:
> ```sql
> %sql
> SELECT student, topic, score
> FROM workspace.genai_labs.sample_table
> WHERE score > 85
> ORDER BY score DESC
> ```
> Both approaches produce identical results.

## Step 6: Clean Up

Good practice: **always clean up test tables** when you are done. Leaving orphan tables in the catalog clutters the schema and confuses collaborators. This is especially important in shared workspaces.

````python
# --------------------------------------------------------------------------
# Step 6 -- Drop the test table to keep the schema clean
# --------------------------------------------------------------------------

spark.sql("DROP TABLE IF EXISTS workspace.genai_labs.sample_table")

print("Cleanup complete: workspace.genai_labs.sample_table has been dropped.")
````

## Exam Tips & Traps

These are the kinds of distinctions the exam loves to test. Read each row carefully.

| # | Topic | Trap Statement | Correct Answer |
|---|---|---|---|
| 1 | Compute | "You must use a cluster for SQL queries" | **WRONG** -- SQL Warehouses are the dedicated compute for SQL workloads. They are separate from interactive clusters and are optimized for BI-style queries. |
| 2 | Serverless | "Serverless requires runtime selection" | **WRONG** -- Serverless manages the runtime automatically. You do not choose a Databricks Runtime version; Databricks picks the appropriate one. |
| 3 | Delta Lake | "Delta tables are just Parquet files" | **WRONG** -- Delta Lake is built *on top of* Parquet, but it adds ACID transactions, versioning (time travel), schema enforcement, and Change Data Feed. These features do not exist in plain Parquet. |
| 4 | Notebooks | "You can only use one language per notebook" | **WRONG** -- Databricks notebooks support magic commands (`%sql`, `%python`, `%scala`, `%r`) to mix languages in the same notebook. |
| 5 | Unity Catalog | "Tables are identified by a two-part name" | **WRONG** -- Unity Catalog uses a three-level namespace: `catalog.schema.table`. The two-part name (`schema.table`) is the legacy Hive metastore convention. |
