## What You Will Build

In this lab, you'll learn about **Agent Bricks** вҖ” Databricks' pre-built agent templates that solve common AI application patterns. Instead of building everything from scratch, Agent Bricks give you a production-ready starting point.

Think of it this way: you *could* build a house from raw lumber, or you could use pre-fabricated walls and just customize the interior. Agent Bricks are the pre-fabricated walls вҖ” they handle the common architecture so you can focus on your specific business logic.

You'll also explore the **Discover** and **Marketplace** pages вҖ” two UI sections where you browse available models, datasets, and pre-built solutions.

### Why This Matters for the Exam

The March 2026 exam added a new objective: *"Determine how and when to use Agent Bricks."* Expect at least one question that describes a business scenario and asks you to pick the correct Agent Brick type. The three types have very distinct use cases вҖ” learn them cold.

## Mental Model

> **Analogy:** Agent Bricks are like LEGO sets. Each set is designed for a specific build:
> - **Knowledge Assistant** = a library set. Shelves (vector search), books (your docs), a librarian (LLM) who answers questions by looking up relevant books. RAG-in-a-box.
> - **AI/BI Genie** = an accountant's desk. It takes natural language questions and translates them into SQL to query your structured tables. Text-to-SQL-in-a-box.
> - **Supervisor Agent** = a construction crew set. A foreman (supervisor) delegates tasks to specialists вҖ” one for plumbing, one for electrical, one for painting. Each sub-agent has its own tools.
> - **Information Extraction** = a data entry set. Takes messy documents (invoices, contracts) and extracts clean structured data (dates, amounts, names) into JSON.
> - **Document Parsing** = an X-ray machine. Understands document *structure* вҖ” tables, layouts, OCR for scanned images.
> - **Custom LLM** = a blank canvas. Specialise an LLM for a custom text task (translation, content generation, code).
> - **Code Your Own Agent** = raw lumber. Build from scratch with LangGraph, DSPy, or other OSS libraries.
>
> Pick the set that matches what you're building.

## Exam Alert

| Trap | Correct Answer |
|---|---|
| "Knowledge Assistant can query structured tables" | **WRONG** вҖ” that's **AI/BI Genie** (Text-to-SQL). Knowledge Assistant does RAG over documents |
| "Knowledge Assistant can orchestrate multiple agents" | **WRONG** вҖ” that's the **Supervisor Agent**. Knowledge Assistant is single-purpose RAG Q&A |
| "Information Extraction generates free-form text answers" | **WRONG** вҖ” it produces *structured* output (JSON, tables) from unstructured documents |
| "Document Parsing and Information Extraction are the same" | **WRONG** вҖ” Document Parsing understands *structure* (tables, layout, OCR). Information Extraction pulls *semantic fields* (names, dates, amounts) |
| "You must build a custom agent for every use case" | **WRONG** вҖ” Agent Bricks provide pre-built templates for the most common patterns |
| "Supervisor Agent handles document extraction" | **WRONG** вҖ” it orchestrates sub-agents. For extraction, use Information Extraction |

> **Exam tip:** Match the *primary goal* to the brick:
> - Q&A over documents вҶ’ **Knowledge Assistant**
> - Q&A over tables/SQL вҶ’ **AI/BI Genie**
> - Route different request types to specialists вҶ’ **Supervisor Agent**
> - Structured data from messy docs вҶ’ **Information Extraction**
> - Understand document layout/tables/OCR вҶ’ **Document Parsing**
> - Custom text task (translate, generate, classify) вҶ’ **Custom LLM**
> - Full control with OSS libraries вҶ’ **Code Your Own Agent**

- **Custom LLM / Code Your Own**: Flexible вҖ” you define the architecture

## Step 3: Business Scenario Matching Practice

This is exactly how the exam tests Agent Bricks вҖ” you get a business scenario and must pick the right type. Let's practice with realistic examples.

```python
# ------------------------------------------------------------------
# Step 3: Interactive scenario matching quiz (all 7 brick types)
# ------------------------------------------------------------------

scenarios = [
    {
        "scenario": "A law firm wants to let lawyers ask questions about case law and internal legal precedents stored in a document repository.",
        "answer": "Knowledge Assistant",
        "why": "This is pure Q&A over a knowledge base (legal documents). RAG pattern."
    },
    {
        "scenario": "An insurance company receives thousands of claim forms (PDFs) daily and needs to extract policyholder name, claim amount, date of incident, and claim type into a database.",
        "answer": "Information Extraction",
        "why": "The goal is to pull structured fields (name, amount, date) from unstructured documents."
    },
    {
        "scenario": "A bank's customer service portal needs to handle billing disputes (-> billing system), loan inquiries (-> loan calculator), and general FAQ (-> knowledge base), routing each to the appropriate handler.",
        "answer": "Supervisor Agent",
        "why": "Different question types need different specialized agents with different tools. The supervisor routes each request."
    },
    {
        "scenario": "A sales team asks 'What was our total revenue by region last quarter?' and expects a data-backed answer from the company's sales Delta tables.",
        "answer": "AI/BI Genie",
        "why": "Natural language question over STRUCTURED tables -> Text-to-SQL. Not docs, not extraction."
    },
    {
        "scenario": "A hospital receives handwritten patient intake forms that have been scanned. They need to extract table rows showing medication names, dosages, and frequencies.",
        "answer": "Document Parsing",
        "why": "Scanned images need OCR + layout analysis to understand table structure. Document Parsing, not Information Extraction."
    },
    {
        "scenario": "A real estate company receives property listing descriptions and needs to extract: address, price, square footage, number of bedrooms into a structured database.",
        "answer": "Information Extraction",
        "why": "Extracting semantic fields from text -> Information Extraction (not Document Parsing, because the input is plain text, not scanned/structured layout)."
    },
    {
        "scenario": "A company wants to automatically translate their English product documentation into Spanish, French, and German.",
        "answer": "Custom LLM",
        "why": "Translation is a custom text task. No retrieval, no extraction, no routing -- just a specialized LLM."
    },
    {
        "scenario": "A startup wants to build a research agent that uses LangGraph to dynamically search the web, read papers, and synthesize findings with custom control flow.",
        "answer": "Code Your Own Agent",
        "why": "Needs full control with a specific OSS library (LangGraph). Pre-built bricks won't give this flexibility."
    },
    {
        "scenario": "An HR department wants employees to ask questions about company policies, benefits, and procedures from the employee handbook.",
        "answer": "Knowledge Assistant",
        "why": "Q&A over a single knowledge base (employee handbook). Simple RAG pattern."
    },
]

print("=" * 70)
print("  AGENT BRICK MATCHING QUIZ")
print("  For each scenario, identify the correct Agent Brick type.")
print("=" * 70)

for i, s in enumerate(scenarios, 1):
    print(f"\n--- Scenario {i} ---")
    print(f"  {s['scenario']}")
    print(f"\n  Answer: {s['answer']}")
    print(f"  Why:    {s['why']}")
```

**What just happened:** You practiced the exact decision pattern the exam tests. The decision rules:

| If the scenario says... | Choose... |
|---|---|
| "Answer questions from documents/knowledge base" | **Knowledge Assistant** |
| "Query structured tables with natural language" | **AI/BI Genie** |
| "Route different request types to different handlers" | **Supervisor Agent** |
| "Extract structured data from unstructured documents" | **Information Extraction** |
| "Parse document layout / OCR / extract tables from images" | **Document Parsing** |
| "Custom text task (translate, generate, classify)" | **Custom LLM** |
| "Build with LangGraph, DSPy, or full OSS control" | **Code Your Own Agent** |

**Tricky distinction:** Document Parsing vs Information Extraction
- **Document Parsing** = understands *physical structure* (where are the tables? what's the layout? OCR the scanned text)
- **Information Extraction** = understands *semantic meaning* (what's the vendor name? what's the amount? classify the sentiment)

## Exam Tips & Traps

**Quick-scan reference -- review these before the exam:**

| # | Topic | Trap | Correct Answer |
|---|---|---|---|
| 1 | Knowledge Assistant vs Genie | "Knowledge Assistant can query tables" | **WRONG** -- Knowledge Assistant searches *documents*. For tables, use **AI/BI Genie** |
| 2 | Supervisor Agent | "A single LLM handles all task types" | **WRONG** -- the supervisor *routes* to specialized sub-agents, each with their own tools |
| 3 | Information Extraction | "Produces free-form text summaries" | **WRONG** -- it produces *structured* data (JSON fields, table rows) |
| 4 | Doc Parsing vs Info Extraction | "They're the same thing" | **WRONG** -- Parsing = physical structure (OCR, layout). Extraction = semantic fields (names, amounts) |
| 5 | Discover vs Marketplace | "They show the same content" | **WRONG** -- Discover = workspace-local assets; Marketplace = installable external assets |
| 6 | Custom agents | "Agent Bricks replace custom agents" | **WRONG** -- Agent Bricks cover common patterns; Code Your Own Agent handles unique requirements |
| 7 | Naming | "Multiagent Supervisor" | The UI calls it **Supervisor Agent** -- same concept, updated name |

> **Pattern to watch for:** The exam describes a scenario and asks "Which Agent Brick should you use?" The answer is always about matching the *primary goal* -- not the industry or company type.

## Key Takeaways

### Core Concepts -- The 7 Agent Bricks
- **Knowledge Assistant** = RAG-in-a-box: Q&A over documents
- **AI/BI Genie** = Text-to-SQL: Q&A over structured tables
- **Supervisor Agent** = orchestrator: routes tasks to specialized sub-agents
- **Information Extraction** = parser: pulls structured JSON from unstructured text
- **Document Parsing** = structure analyzer: OCR, layout analysis, table extraction
- **Custom LLM** = specialized model: translation, content generation, code
- **Code Your Own Agent** = full OSS control: LangGraph, DSPy, any framework

### Databricks-Specific
- **UI ->** Left nav -> **Agents** -> **Agent Bricks** -> **New** to see all 7 types
- **UI ->** Left nav -> **Discover** to browse models and datasets in your workspace
- **UI ->** Left nav -> **Marketplace** to browse and install external assets
- Foundation models have a `databricks-` prefix and are pre-hosted (zero setup)

### Exam Essentials
- Know all seven Agent Brick types and their exact use cases
- The decision is based on the *primary goal*, not the industry
- **Docs** -> Knowledge Assistant | **Tables** -> AI/BI Genie | **Route** -> Supervisor Agent
- **Extract fields** -> Information Extraction | **OCR/layout** -> Document Parsing
- **Custom text** -> Custom LLM | **Full control** -> Code Your Own Agent
