## What You Will Build

In this lab, you'll learn about **Agent Bricks** — Databricks' pre-built agent templates that solve common AI application patterns. Instead of building everything from scratch, Agent Bricks give you a production-ready starting point.

Think of it this way: you *could* build a house from raw lumber, or you could use pre-fabricated walls and just customize the interior. Agent Bricks are the pre-fabricated walls — they handle the common architecture so you can focus on your specific business logic.

You'll also explore the **Discover** and **Marketplace** pages — two UI sections where you browse available models, datasets, and pre-built solutions.

### Why This Matters for the Exam

The March 2026 exam added a new objective: *"Determine how and when to use Agent Bricks."* Expect at least one question that describes a business scenario and asks you to pick the correct Agent Brick type. The three types have very distinct use cases — learn them cold.

## Mental Model

> **Analogy:** Agent Bricks are like LEGO sets. Each set is designed for a specific build:
> - **Knowledge Assistant** = a library set. It comes with shelves (vector search), books (your documents), and a librarian (the LLM) who answers questions by looking up relevant books. This is RAG-in-a-box.
> - **Multiagent Supervisor** = a construction crew set. It has a foreman (supervisor agent) who delegates tasks to specialists — one for plumbing, one for electrical, one for painting. Each sub-agent has its own tools.
> - **Information Extraction** = a data entry set. It takes messy, unstructured documents (invoices, contracts, emails) and extracts clean, structured data (dates, amounts, names) into a table format.
>
> Pick the set that matches what you're building.

## Exam Alert

| Trap | Correct Answer |
|---|---|
| "Knowledge Assistant can orchestrate multiple agents" | **WRONG** — that's the Multiagent Supervisor. Knowledge Assistant is single-purpose RAG Q&A |
| "Information Extraction generates free-form text answers" | **WRONG** — it produces *structured* output (JSON, tables) from unstructured documents |
| "You must build a custom agent for every use case" | **WRONG** — Agent Bricks provide pre-built templates for the three most common patterns |
| "Multiagent Supervisor handles document extraction" | **WRONG** — it orchestrates sub-agents. For extraction, use Information Extraction |

> **Exam tip:** When a scenario describes answering questions from a knowledge base → **Knowledge Assistant**. When it describes routing different question types to different specialists → **Multiagent Supervisor**. When it describes pulling structured data from messy documents → **Information Extraction**.

## Step 3: Business Scenario Matching Practice

This is exactly how the exam tests Agent Bricks — you get a business scenario and must pick the right type. Let's practice with realistic examples.

```python
# ------------------------------------------------------------------
# Step 3: Interactive scenario matching quiz
# ------------------------------------------------------------------

scenarios = [
    {
        "scenario": "A law firm wants to let lawyers ask questions about case law and internal legal precedents stored in a document repository.",
        "answer": "Knowledge Assistant",
        "why": "This is pure Q&A over a knowledge base (legal documents). No multi-agent routing needed, no structured extraction needed."
    },
    {
        "scenario": "An insurance company receives thousands of claim forms (PDFs) daily and needs to extract policyholder name, claim amount, date of incident, and claim type into a database.",
        "answer": "Information Extraction",
        "why": "The goal is to pull structured fields from unstructured documents. This is the textbook IE use case."
    },
    {
        "scenario": "A bank's customer service portal needs to handle billing disputes (→ billing system), loan inquiries (→ loan calculator), and general FAQ (→ knowledge base), routing each to the appropriate handler.",
        "answer": "Multiagent Supervisor",
        "why": "Different question types need different specialized agents with different tools. The supervisor routes each request."
    },
    {
        "scenario": "An HR department wants employees to ask questions about company policies, benefits, and procedures from the employee handbook.",
        "answer": "Knowledge Assistant",
        "why": "Q&A over a single knowledge base (employee handbook). Simple RAG pattern."
    },
    {
        "scenario": "A real estate company receives property listing descriptions and needs to extract: address, price, square footage, number of bedrooms, and listing date into a structured database.",
        "answer": "Information Extraction",
        "why": "Extracting structured fields from unstructured text → IE."
    },
    {
        "scenario": "A manufacturing company needs an agent that can: check inventory (→ inventory system), place orders (→ procurement system), and answer questions about product specs (→ technical docs).",
        "answer": "Multiagent Supervisor",
        "why": "Three distinct capabilities requiring different tools and data sources. Supervisor coordinates specialized sub-agents."
    }
]

print("=" * 70)
print("  AGENT BRICK MATCHING QUIZ")
print("  For each scenario, identify the correct Agent Brick type.")
print("=" * 70)

for i, s in enumerate(scenarios, 1):
    print(f"\n--- Scenario {i} ---")
    print(f"  {s['scenario']}")
    print(f"\n  ✅ Answer: {s['answer']}")
    print(f"  💡 Why:    {s['why']}")
```

**What just happened:** You practiced the exact decision pattern the exam tests. The decision rules are simple:

| If the scenario says... | Choose... |
|---|---|
| "Answer questions from documents/knowledge base" | **Knowledge Assistant** |
| "Route different types of requests to different handlers" | **Multiagent Supervisor** |
| "Extract structured data from unstructured documents" | **Information Extraction** |

## Exam Tips & Traps

**Quick-scan reference — review these before the exam:**

| # | Topic | Trap | Correct Answer |
|---|---|---|---|
| 1 | Knowledge Assistant | "Can handle multiple specialized tasks" | **WRONG** — it's single-purpose RAG Q&A. For multiple tasks, use Multiagent Supervisor |
| 2 | Multiagent Supervisor | "A single LLM handles all task types" | **WRONG** — the supervisor *routes* to specialized sub-agents, each with their own tools |
| 3 | Information Extraction | "Produces free-form text summaries" | **WRONG** — it produces *structured* data (JSON fields, table rows) |
| 4 | Discover vs Marketplace | "They show the same content" | **WRONG** — Discover = workspace-local assets; Marketplace = installable external assets |
| 5 | Custom agents | "Agent Bricks replace custom agents" | **WRONG** — Agent Bricks cover common patterns; custom agents handle unique requirements |

> **Pattern to watch for:** The exam describes a scenario and asks "Which Agent Brick should you use?" The answer is always about matching the *primary goal* — not the industry or company type.

## Key Takeaways

### Core Concepts
- **Knowledge Assistant** = RAG-in-a-box: Q&A over a document knowledge base
- **Multiagent Supervisor** = orchestrator: routes different tasks to specialized sub-agents
- **Information Extraction** = parser: pulls structured data from unstructured documents
- Agent Bricks are starting points, not final products — you customize them for your use case

### Databricks-Specific
- **UI ->** Left nav -> **Discover** to browse models and datasets in your workspace
- **UI ->** Left nav -> **Marketplace** to browse and install external models and datasets
- Foundation models have a `databricks-` prefix and are pre-hosted (zero setup)
- Agent Bricks integrate with Unity Catalog, MLflow, and Model Serving

### Exam Essentials
- Know all three Agent Brick types and their exact use cases
- The decision is based on the *primary goal*, not the industry
- Q&A from docs → Knowledge Assistant
- Multiple task types with routing → Multiagent Supervisor
- Structured data from unstructured docs → Information Extraction
